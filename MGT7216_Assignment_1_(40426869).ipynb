# Integration of Cleaned Data and Segregation of Entries for Update

**Step 1: Creating a Subset for Entries Needing Update**

First, identify and separate entries with 'NaN' or 'U'. This subset will be used for potential user updates or further review.

```python
needs_update = df[(df['gender'] == 'U') |
                  (df['fam_income'] == 'U') |
                  (df['child'] == 'U') |
                  df['education'].isnull() |
                  df['marriage'].isnull() |
                  df['house_owner'].isnull() |
                  (df['age'] == '1_Unk')]

# Excluding these entries from the df_cleaned dataset
df_cleaned = df.drop(needs_update.index)

# Further cleaning steps, if necessary, on df_cleaned
df_cleaned['child'] = df_cleaned['child'].replace({'0': 'N'}).replace({'U': np.nan})
df_cleaned = df_cleaned.dropna()  # This will drop rows where any column has 'NaN'

# Saving the 'needs_update' subset for later user engagement or correction
needs_update.to_csv('needs_update.csv', index=False)
needs_update
```

**Step 2: Merging Cleaned Data Back to the Main Dataset**

Next, integrate the cleaned data back into the main dataset, excluding those entries flagged for update. This involves creating a cleaned version of your dataset by excluding problematic entries identified above.

```python
# Exclude entries identified for updates from the main dataset
df_useable = df.copy().drop(needs_update.index)

# Perform cleaning steps on the rest of the dataset
df_useable['child'] = df_useable['child'].replace({'0': 'N'}).replace({'U': np.nan})
df_useable = df_useable.dropna(subset=['education', 'marriage', 'house_owner', 'age', 'child', 'gender', 'fam_income'])
df_useable
```

This code effectively splits your dataset into two: 'df_useable', which excludes entries with 'NaN' and 'U' and contains cleaned data ready for analysis, and needs_update, a subset of entries identified for potential correction. This approach ensures that your analysis is based on clean, reliable data, while also retaining a path for improving data completeness and accuracy by addressing the entries in needs_update.

**Enhancing Data Analysis through Numerical Encoding**

**Objective:**

To unlock a broader spectrum of statistical analyses and derive more profound insights from our dataset, I transformed several categorical variables into numerical formats. This approach was applied to 'education', 'fam_income', 'child', 'online', and 'house_owner'. The decision to maintain the categorical representation of 'age' stems from its descriptive nature, which provides significant value in its current format.

**Theoretical Justification:**

The practice of converting ordinal and specific nominal variables to numerical formats is well-established in data science, enhancing the analytical utility of ordinal data by facilitating quantitative analyses. Literature such as Agresti (2010) shows the advantages of numerically treating ordinal variables for applying sophisticated statistical models. Stevens (1946) further supports this approach by highlighting how ordinal data can be quantitatively interpreted to reflect the inherent order of categories, thus enabling a wider array of statistical tests and analyses.

Additionally, we are not deleting the original variables, but rather creating new ones alongside the originals for further analysis.

**Implementation Approach:**

*'education' and 'fam_income':*

These variables are inherently ordinal. By converting them into a numeric scale, we adhere to best practices allowing for ordinal logistic regression, correlation analysis, and more.

*'child', 'online', and 'house_owner":*

This binary variable (Yes/No) is straightforward to convert into numeric format (1/0), facilitating its inclusion in various quantitative analyses.

```python
# Mapping 'education' categories to numeric values
education_mapping = {'0. <HS': 0, '1. HS': 1, '2. Some College': 2, '3. Bach': 3, '4. Grad': 4}
df_cleaned['education_numeric'] = df_cleaned['education'].map(education_mapping)

# Mapping 'fam_income' categories to numeric values, assuming a logical order from low to high
income_mapping = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7, 'H': 8, 'I': 9, 'J': 10, 'K': 11, 'L': 12}
df_cleaned['fam_income_numeric'] = df_cleaned['fam_income'].map(income_mapping)

# Converting 'child' to numeric values ('Y' to 1 and 'N' to 0)
child_mapping = {'Y': 1, 'N': 0}
df_cleaned['child_numeric'] = df_cleaned['child'].map(child_mapping)

# Convert 'online' to numeric: 'Y' to 1 and 'N' to 0
df_cleaned['online_numeric'] = df_cleaned['online'].map({'Y': 1, 'N': 0})

# Convert 'house_owner' to numeric: 'Owner' to 1 and 'Renter' to 0
df_cleaned['house_owner_numeric'] = df_cleaned['house_owner'].map({'Owner': 1, 'Renter': 0})

# Verify the conversion
print(df_cleaned[['education_numeric', 'fam_income_numeric', 'child_numeric', 'online_numeric', 'house_owner_numeric']].head())
```

This section documents the transformation of categorical variables into a numeric format, enhancing the dataset's suitability for a wider array of analytical techniques. By selecting variables for conversion based on their ordinal nature and relevance to our research questions, we adhere to methodological best practices. This process not only aligns with the literature on data analysis strategies but also significantly broadens the scope of potential insights derivable from our dataset, ensuring that our findings are both robust and deeply informed by the underlying data structure.

**Splitting the Dataset for Model Training**

**Preparing for Predictive Modeling**

**Objective:**
To build a robust predictive model, the first essential step is to split our dataset into two distinct sets: a training set, which will be used to teach the model about the patterns that lead to a purchase, and a test set, to evaluate the model's predictive accuracy on data it hasn't seen before. This ensures we can measure the model's performance fairly and reliably.

```python
from sklearn.model_selection import train_test_split

# 'flag' is our target variable indicating whether a customer has purchased a product.
X = df_cleaned.drop(['flag'], axis=1)  # Exclude the target variable to isolate features
y = df_cleaned['flag'].map({'Y': 1, 'N': 0})  # Convert the target variable to a binary numeric format, with 'Y' as 1 and 'N' as 0

# Splitting the data into training (80%) and test sets (20%).
# I chose a 'random_state' to ensure our results are reproducible.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40426869)

print(f"Training Features Shape: {X_train.shape}")
print(f"Training Labels Shape: {y_train.shape}")
print(f"Test Features Shape: {X_test.shape}")
print(f"Test Labels Shape: {y_test.shape}")
```

# Tune One or Two Parameters

**Hyperparameter Tuning for the Logistic Regression Model**

**Objective:**
The aim is to optimise key hyperparameters of the Logistic Regression model to improve its predictive accuracy and generalisation to unseen data. Logistic Regression has fewer hyperparameters compared to Random Forest, but we can still explore tuning parameters like C (inverse of regularisation strength) and penalty (type of regularisation) to enhance model performance.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define the parameter grid to search over for Logistic Regression
param_grid_lr = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Regularisation strength
    'penalty': ['l1', 'l2'],  # Regularisation type
    'solver': ['liblinear']  # Algorithm to use in the optimization problem, suitable for small datasets and 'l1' penalty
}

# Initialise the GridSearchCV object for Logistic Regression
grid_search_lr = GridSearchCV(estimator=LogisticRegression(random_state=40426869, max_iter=10000),
                              param_grid=param_grid_lr,
                              cv=5,  # 5-fold cross-validation
                              scoring='accuracy',  # Evaluation metric
                              verbose=1)  # Show progress

# Fit the GridSearchCV object to the encoded training data
grid_search_lr.fit(X_train_encoded, y_train)

# Print the best parameters and the best score for Logistic Regression
print("Logistic Regression - Best Parameters:", grid_search_lr.best_params_)
print("Logistic Regression - Best Score:", grid_search_lr.best_score_)

# Use the best estimator for further predictions and evaluations for Logistic Regression
best_lr_model = grid_search_lr.best_estimator_
```

**Optimised Logistic Regression Model Insights**

After tuning, the Logistic Regression model's optimal parameters are {'C': 10, 'penalty': 'l1', 'solver': 'liblinear'}, achieving a best cross-validation score of 69.93%. This improvement highlights the benefits of fine-tuning, where a lower regularisation strength and L1 penalty enhance the model's accuracy and potentially its interpretability by selecting more relevant features. These results underscore the effectiveness of hyperparameter tuning in boosting model performance.

# Comprehensive Model Evaluation: A Comparative Analysis of Predictive Performance

**Comparative Analysis and Literature Link**

Both the Logistic Regression and Random Forest models, post-optimisation, present viable options for enhancing Imperials Ltd's predictive marketing capabilities. The choice between them may depend on specific preferences for simplicity and interpretability (Logistic Regression) versus a slight edge in performance and depth of insight (Random Forest). Literature supports the evaluation of multiple models to identify the best fit for specific data and business needs, emphasising the importance of balanced precision and recall across classes in predictive modeling.

**Explanation and Literature Link:**

In this section, we delve into Receiver Operating Characteristic (ROC) curve analysis to assess and compare the discriminatory power of our Logistic Regression and Random Forest models. ROC curves plot the true positive rate (TPR) against the false positive rate (FPR) at various threshold settings, providing a comprehensive measure of model performance across different classification thresholds. The Area Under the Curve (AUC) serves as an aggregate measure of the model's ability to correctly classify positive and negative instances, with values closer to 1 indicating superior predictive performance.

ROC curve analysis is a widely recognised method for evaluating the predictive accuracy of classification models, offering insights beyond what is available through accuracy alone. This approach is supported by Fawcett's (2006) work, "An Introduction to ROC Analysis," which highlights the value of ROC curves in comparing classifiers across diverse decision thresholds, making it particularly relevant for Imperials Ltd's need to balance identifying potential customers with minimising false marketing outreach.

```python
# Generating probability predictions for the Logistic Regression model
y_test_pred_proba_lr = best_lr_model.predict_proba(X_test_encoded)[:, 1]

# Generating probability predictions for the Random Forest model
y_test_pred_proba_rf = best_rf_model.predict_proba(X_test_encoded)[:, 1]

from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Compute ROC curve and AUC for Logistic Regression
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_test_pred_proba_lr)
roc_auc_lr = auc(fpr_lr, tpr_lr)

# Compute ROC curve and AUC for Random Forest
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_test_pred_proba_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)

# Plot ROC curves
plt.figure(figsize=(10, 8))
plt.plot(fpr_lr, tpr_lr, label=f'Optimised Logistic Regression (AUC = {roc_auc_lr:.2f})')
plt.plot(fpr_rf, tpr_rf, label=f'Optimised Random Forest (AUC = {roc_auc_rf:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random Chance')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Optimised Models')
plt.legend(loc='lower right')
plt.show()
```

ROC curves for the optimised Logistic Regression and Random Forest models have been successfully generated. The AUC (Area Under the Curve) scores for both models are quite close, with the Random Forest model having a slight edge:

**Optimised Logistic Regression (AUC = 0.74):** This indicates that the model has a good ability to distinguish between the positive and negative classes. An AUC of 0.74 means the Logistic Regression model correctly discriminates between a customer who will buy and who wonâ€™t buy 74% of the time.

**Optimised Random Forest (AUC = 0.75):** A slightly better AUC score than Logistic Regression implies that the Random Forest model is slightly more effective at distinguishing between the two classes. An AUC of 0.75 means the Random Forest model has a 75% chance of correctly distinguishing between a customer who will buy and one who will not.

Both models perform significantly better than random chance, which would have an AUC of 0.5 (represented by the dashed line). The closeness of their AUC scores suggests that either model could be useful for Imperials Ltd, with the choice potentially coming down to other factors such as interpretability, training time, or model complexity.

```python
# Enhanced hyperparameter grid for Logistic Regression
param_grid_lr_advanced = {
    'C': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000],
    'penalty': ['l1', 'l2', 'elasticnet'],
    'solver': ['liblinear', 'saga'],
    'l1_ratio': [0, 0.5, 1]
}

# Grid search to find the best parameters for Logistic Regression
grid_search_lr_advanced = GridSearchCV(
    estimator=LogisticRegression(random_state=40426869, max_iter=10000),
    param_grid=param_grid_lr_advanced,
    cv=5,
    scoring='accuracy',
    verbose=1
)

# Fit the model on the training data
grid_search_lr_advanced.fit(X_train_encoded, y_train)
```

**Further Tuning of the Random Forest Model:**

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint

# Define the parameter distribution to sample from for Random Forest
param_dist_rf = {
    'n_estimators': randint(100, 500),  # Randomly sample number of trees
    'max_depth': [10, 20, 30, 40, None],  # Depth of trees
    'min_samples_split': randint(2, 11),  # Randomly sample for minimum split
    'min_samples_leaf': randint(1, 5),  # Randomly sample for minimum leaf
    'max_features': ['auto', 'sqrt', 'log2']  # The number of features to consider
}

# Initialize the RandomizedSearchCV object for Random Forest
random_search_rf = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=40426869),
    param_distributions=param_dist_rf,
    n_iter=100,  # Number of parameter settings that are sampled
    cv=5,  # 5-fold cross-validation
    scoring='accuracy',  # Evaluation metric
    verbose=1,  # Show progress
    random_state=40426869,
    n_jobs=-1  # Use all available cores
)

# Fit the RandomizedSearchCV object to the encoded training data
random_search_rf.fit(X_train_encoded, y_train)

# Print the best parameters and the best score for Random Forest
print("Random Forest - Best Parameters:", random_search_rf.best_params_)
print("Random Forest - Best Score:", random_search_rf.best_score_)

# Use the best estimator for further predictions and evaluations
best_rf_model_advanced = random_search_rf.best_estimator_
```
